```python
from datasets import load_dataset
# Load WikiText-2 raw
dataset = load_dataset("wikitext", "wikitext-2-raw-v1")
print(dataset)
```

```python
for i in range(10):
    print(i)
    print(dataset["train"][i]["text"][:120])
```

```python
print(len(dataset["train"]))
```

```python
from collections import Counter
all_text = " ".join(dataset["train"]["text"])
tokens = all_text.split()
print(f"Total tokens : {len(tokens):,}")
print(f"Vocab size   : {len(set(tokens)):,}")
# Token frequency
freq = Counter(tokens).most_common(20)
for tok, cnt in freq:
    print(f"  {tok:10s} {cnt:>7,}")
# Sentence length distribution
import matplotlib.pyplot as plt
lengths = [len(t.split()) for t in dataset["train"]["text"] if t.strip()]
plt.hist(lengths, bins=50, edgecolor="black")
plt.xlabel("Tokens per line")
plt.ylabel("Count")
plt.title("WikiText-2 Line Length Distribution")
plt.show()
```
